## Útil

### **Modelo de dados**
(ou Modelo de objetos)

> "...uma descrição de Python na forma de um framework. Ele formaliza as interfaces dos blocos de construção da própria linguagem, por exemplo, as sequências, os iteradores, as funções, as classes, os gerenciadores de contexto e assim por diante."

**Objeto** é a abstração que o Python usa para se referir a qualquer tipo de dado e todos eles são representados por objetos ou pelo relacionamento entre objetos. Sendo que cada um tem uma identidade (que nunca muda), tipo e valor.




### **Métodos especiais**

Permitem que os seus objetos implementem, tratem e interajam com as construções básicas da linguagem (iterações, coleções, chamadas de funções, etc.)

O interpretador Python chama métodos especiais para realizar operações básicas em objetos, geralmente acionados por uma sintaxe especial. Pense em qualquer operação comum... provavelmente ela já está criada.

image.png

Consulta: https://speakerdeck.com/ramalho/a-beleza-de-python?slide=55

- A sintaxe `obj[key]` é tratada pelo método especial `__getitem__` (chamado também como método mágico ou método dunder). Para avaliar `my_collection[key]`, o interpretador chama `my_collection.__getitem__(key)`.

- A chamada deve acontecer pelo interpretador Python, e não por nós. Não escrevemos my_object.__len__() e sim `len(my_object)`. Se `my_object` for uma instância de uma classe definida, o Python chamará o método de instância `__len__` implementado.

- A implementação de métodos especiais é mais comum que chamá-los diretamente, a não ser que esteja utilizando **metaprogramação**.

- A exceção para o último item é o `__init__` (iniciador da classe, para invocar o inicializador da superclasse quando há a implementação direta).

- De forma geral, é melhor **chamar a função embutida relacionada**, pois são mais rápidas.



## Um baralho pythônico & Métodos Especiais

```python
import collections

Card = collections.namedtuple('Card', ['rank', 'suit'])

class FrenchDeck:
    ranks = [str(n) for n in range(2, 11)] + list('JQKA')
    suits = 'spades diamonds clubs hearts'.split()

    def __init__(self):
        self._cards = [
            Card(rank, suit)
            for suit in self.suits
            for rank in self.ranks
        ]

    def __len__(self):
        return len(self._cards)

    def __getitem__(self, position):
        return self._cards[position]
```

**namedtuple**: Criar classes de objetos que sejam apenas grupos de atributos, sem métodos próprios.

```python
def spades_high(card):
    rank_value = FrenchDeck.ranks.index(card.rank)
    return rank_value * len(suit_values) + suit_values[card.suit]
```

## Tipos numéricos com vetores

```python
import math

class Vector:

    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def __repr__(self):
        return f'Vector({self.x!r}, {self.y!r})'

    def __abs__(self):
        return math.hypot(self.x, self.y)

    def __bool__(self):
        return bool(abs(self))

    def __add__(self, other):
        x = self.x + other.x
        y = self.y + other.y
        return Vector(x, y)

    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)

    def __repr__(self):
        return f'Vector({self.x}, {self.y})'
    
    def __bool__(self):
        return self.x or self.y
```

```python
math.hypot(3, 4)
```

```python
v1 = Vector(2, 4)
v2 = Vector(4, 5)

v1 + v2
```

## Representação em string
Diferença entre os métodos especiais __str__ and __repr__.

```python
class Hi(object):
    pass

print(str(Hi()))
print(repr(Hi()))
```

```python
class Hi(object):
    def __repr__(object):
        return 'Hi'

print(str(Hi()))
print(repr(Hi()))
```

```python
class Hi(object):
    def __str__(object):
        return 'Hi'

print(str(Hi()))
print(repr(Hi()))
```

```python
m = 'Mari'
repr(m)
```

```python
str(m)
```

```python
eval(repr(m)) == m
```

## Valores booleanos
Por padrão, as instâncias das classes definidas pelo usuário são consideradas verdadeiras (a menos que __bool__ ou __len__ esteja implementado).

Se não tiver __bool__ o Python chama o __len__, se igual a 0, bool retorna False.
